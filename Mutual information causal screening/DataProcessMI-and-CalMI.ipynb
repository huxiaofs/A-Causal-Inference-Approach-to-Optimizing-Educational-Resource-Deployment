```python
# 计算离散变量之间的互信息
for var in discrete_vars:
    print(var)
    mi_score = mutual_info_classif(final_df[[var]], final_df[discreted_target_var])
    mi_results[var] = mi_score
```

```python
for var in continuous_vars:
    
    mi_score = mutual_info_regression(final_df[[var]], final_df[target_var])
    mi_results[var] = mi_score
    print(var, mi_score)
```
